## –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

### 1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.1.
### 2. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  ### 3. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')

f = 10*x - 5
df = sp.diff(f, x)

#–ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å –∑ —Å–∏–º–≤–æ–ª—å–Ω–∏—Ö —É —Ñ—É–Ω–∫—Ü—ñ—ó
f_func = sp.lambdify(x, f, 'numpy')
df_value = float(df)
x_vals = np.linspace(0, 10, 100)  # –î—ñ–∞–ø–∞–∑–æ–Ω –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω
y_vals = f_func(x_vals)
dy_vals = np.full_like(x_vals, df_value)  # –ú–∞—Å–∏–≤ –∑—ñ –∑–Ω–∞—á–µ–Ω—å 10


#–°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = 10ùë•‚àí5', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = 10", color='green')


plt.title('–ó–∞–ª–µ–∂–Ω—ñ—Å—Ç—å –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è')
plt.xlabel('–ì–æ–¥–∏–Ω–∏ –Ω–∞–≤—á–∞–Ω–Ω—è')
plt.ylabel('–û—Ü—ñ–Ω–∫–∞')


plt.show()
```

###–≤–∏—Å–Ω–æ–≤–æ–∫
1. –æ—Å–∫—ñ–ª—å–∫–∏ –ø–æ—Ö—ñ–¥–Ω–∞ –¥–æ–¥–∞—Ç–Ω—è —Ç–æ –æ—Ü—ñ–Ω–∫–∞ –∑—Ä–æ—Å—Ç–∞—î —Å—Ç–∞–±—ñ–ª—å–Ω–æ –≤—ñ–¥ –∫—ñ–ª-—Ç—ñ –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –≥–æ–¥–∏–Ω
2. —Å–∏–Ω—è –ª—ñ–Ω—ñ—è –Ω–∞—à–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è —Å—Ç—Ä—ñ–º–∫–æ –∑—Ä–æ—Å—Ç–∞—î –≤–≥–æ—Ä—É –ø–æ–∫–∞–∑—É—é—á–∏ –Ω–∞–º —â–æ, –≤—ñ–¥ –∫—ñ–ª-—Ç—ñ –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è —Å—Ç—Ä—ñ–º–∫–æ –∑—Ä–æ—Å—Ç–∞—î –æ—Ü—ñ–Ω–∫–∞
3. –æ—Å–∫—ñ–ª—å–∫–∏ –ø–æ—Ö—ñ–¥–Ω–∞ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∞ –∑—ñ –∑–Ω–∞—á–µ–Ω–Ω—è–º 10 —Ç–æ –∑ –∫–æ–∂–Ω–æ—é –≥–æ–¥–∏–Ω–æ—é –Ω–∞–≤—á–∞–Ω–Ω—è –æ—Ü—ñ–Ω–∫–∞ –∑—Ä–æ—Å—Ç–∞—î –Ω–∞ 10 –±–∞–ª—ñ–≤ –Ω–µ –∑–∞–ª–µ–∂–Ω–æ —Ü–µ –ø–µ—Ä—à–∞ —á–∏ 5—Ç–∞ –≥–æ–¥–∏–Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è

###2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



#ùëì(ùë•)=sin(ùë•)

```python

x = sp.symbols('x')

f = sp.sin(x)
df = sp.diff(f, x)

#–ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å –∑ —Å–∏–º–≤–æ–ª—å–Ω–∏—Ö —É —Ñ—É–Ω–∫—Ü—ñ—ó
f_func = sp.lambdify(x, f, 'numpy')
df_func = sp.lambdify(x, df, 'numpy')
x_vals = np.linspace(- 4*np.pi, 4 * np.pi, 400)  # –î—ñ–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω—å —Ö
y_vals = f_func(x_vals)
dy_vals = df_func(x_vals)


#–°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = cos(x)", color='green')
plt.title("–°–∏–Ω—É—Å —Ç–∞ –π–æ–≥–æ –ø–æ—Ö—ñ–¥–Ω–∞")


plt.show()
```

#ùëì(ùë•)=ln(ùë•)

```python
x = sp.symbols('x')

f = sp.ln(x)
df = sp.diff(f, x)

#–ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å –∑ —Å–∏–º–≤–æ–ª—å–Ω–∏—Ö —É —Ñ—É–Ω–∫—Ü—ñ—ó
f_func = sp.lambdify(x, f, 'numpy')
df_func = sp.lambdify(x, df, 'numpy')
x_vals = np.linspace(0.1,10,50)  # –î—ñ–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω—å —Ö
y_vals = f_func(x_vals)
dy_vals = df_func(x_vals)


#–°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = 1/x", color='green')
plt.title("ln —Ç–∞ –π–æ–≥–æ –ø–æ—Ö—ñ–¥–Ω–∞")


plt.show()
```


###3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
x = sp.symbols('x')

f = x**2-5*x+1
df = sp.diff(f, x)

#–ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å –∑ —Å–∏–º–≤–æ–ª—å–Ω–∏—Ö —É —Ñ—É–Ω–∫—Ü—ñ—ó
f_func = sp.lambdify(x, f, 'numpy')
df_func = sp.lambdify(x, df, 'numpy')
x_vals = np.linspace(0,100,100000)  # –î—ñ–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω—å —Ö
y_vals = f_func(x_vals)
dy_vals = df_func(x_vals)


#–°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = 1/x", color='green')
plt.title("  —Ñ—É–Ω–∫—Ü—ñ—è —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞")


plt.show()
```

–Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É –≤–≥–æ—Ä—ñ –º–∞–π–∂–µ –Ω–µ –ø–æ–º—ñ—Ç–Ω–æ, —â–æ —î –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –≤—ñ–¥ –∑–Ω–∞—á–µ–Ω–Ω—è —Ö, –æ—Å–∫—ñ–ª—å–∫–∏ —Å–∞–º–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑—Ä–æ—Å—Ç–∞—î —à–≤–∏–¥—à–µ –Ω—ñ–∂ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞. –Ω–∏–∂—á–µ –Ω–∞–º–∞–ª—å–æ–≤–∞–Ω–∏–π –≥—Ä–∞—Ñ—ñ–∫ –∑ —Ö –≤—ñ–¥ 0 –¥–æ 10 —ñ –Ω–∞ —Ü—å–æ–º—É –≥—Ä–∞—Ñ—ñ–∫—É –≥–∞—Ä–Ω–æ –≤–∏–¥–Ω–æ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è –ø–æ—Ö—ñ–¥–Ω–æ—ó –∑ —Ä–æ—Å—Ç–æ–º —Ö

```python
x = sp.symbols('x')

f = x**2-5*x+1
df = sp.diff(f, x)

#–ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å –∑ —Å–∏–º–≤–æ–ª—å–Ω–∏—Ö —É —Ñ—É–Ω–∫—Ü—ñ—ó
f_func = sp.lambdify(x, f, 'numpy')
df_func = sp.lambdify(x, df, 'numpy')
x_vals = np.linspace(0,10,100)  # –î—ñ–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω—å —Ö
y_vals = f_func(x_vals)
dy_vals = df_func(x_vals)


#–°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals, dy_vals, label="f'(x) = 1/x", color='green')
plt.title("  —Ñ—É–Ω–∫—Ü—ñ—è —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞")


plt.show()
```
