```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
#1
result_matrix =np.dot(  B,A)
display(result_matrix)
```

```python
#2
result_C = np.dot(result_matrix,C)
display(result_C)
```

```python
#3
result_D=np.dot(result_matrix,D)
display(result_D)
```
